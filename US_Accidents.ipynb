Import Necessary Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Read the dataset

```python
df = pd.read_csv('US_Accidents_March23.csv')
df
```

```python
df.info()
```

```python
df.columns
```


Removing Unneccesary Columns

```python
col = ['End_Lat','End_Lng','Description','Street','County','Zipcode','Airport_Code','Weather_Timestamp','Wind_Chill(F)','Wind_Direction','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight']
```

```python
df=df.drop(col, axis=1)
df.columns
```

```python
df['Source'].unique()
```


Analyze Accident Count of cities

```python
df_city=df.City.value_counts()
df_city
```

```python
df_city[:20].plot(kind='barh')
plt.title("Accident Count of top 20 Cities")
plt.show()
```


Visualizing various features of the dataset

```python
df1=['Amenity', 'Crossing','Give_Way', 'Junction', 'No_Exit', 'Railway', 'Station','Stop', 'Traffic_Calming', 'Traffic_Signal','Sunrise_Sunset']
```

```python
fig,ax=plt.subplots(3,4,figsize=(20,30))
for i in range(len(df1)):
    df_plot = df.groupby([df1[i], "Severity"]).size().reset_index().pivot(columns=df1[i], index="Severity", values=0)
    df_plot.div(df_plot.sum(axis=1), axis=0).plot(kind='bar', stacked=True,ax=ax[int(i/4),i-int(i/4)*4])
display()
```


Temperature(F) Analysis

```python
x = df['Temperature(F)']
bins = np.arange(df['Temperature(F)'].min(), df['Temperature(F)'].max(),20.0)
plt.hist(x, bins, alpha=0.8, histtype='bar', ec='black')
plt.xlabel('Temperature(F)')
plt.ylabel('count')
plt.xticks(bins)
plt.show()
```

The plot shows that most accidents occured when temperature was between 50-90 Fahrenheit.

```python
df_source=df['Source'].value_counts()
df_source.plot(kind='bar')
plt.title("Accident Count of different source")
plt.show()
```

```python

bins = np.arange(df['Humidity(%)'].min(), df['Humidity(%)'].max(),5.0)
plt.hist(df['Humidity(%)'], bins, alpha=0.8, histtype='bar', ec='black')
plt.xlabel('Humidity(%)')
plt.ylabel('Count')
plt.xticks(bins)
plt.show()
```

This plot shows that high humidity may lead into the accident and increases risks for human life.


```python
bins = np.arange(df['Visibility(mi)'].min(), df['Visibility(mi)'].max(),5.0)
plt.hist(df['Visibility(mi)'], bins, alpha=0.8, histtype='bar', ec='black')
plt.xlabel('Visibility(mi)')
plt.ylabel('Count')
plt.xticks(bins)
plt.show()
```

This plot conveys the visibility issues leading to most number of accidents.




Visualizing Severity of Accidents in various coordinates

```python
fig,ax=plt.subplots(1,figsize=(20,10))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=df, hue='Severity',palette='RdYlGn_r',ax=ax)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Severity of accidents by their coordinates')
plt.show()
```

This plot conveys that how much an area is prone to the accident, which advises us to be more careful while driving through them.
